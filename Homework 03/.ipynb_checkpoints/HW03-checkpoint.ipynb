### 1.Loading the data from multiple excel files into the data frame

```python
all_data = pd.DataFrame()
import glob
for f in glob.glob("input/*.xls*"):
    df = pd.read_excel(f)
    
    #substring student id
    idstart = f.find('6')
    idend = f.find('.')
    id = f[idstart:idend]
    
    #append column id before add into all_data
    df['Id'] = id
    
    all_data = all_data.append(df,ignore_index=True)
```
